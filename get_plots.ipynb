```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
```

```python
results_dir = 'results/'
```

```python
# train loss
train_rnn_loss_1 = genfromtxt("results/simplernn_train_loss_run1.txt", delimiter=',')[:100]
# test loss
test_rnn_loss_1 = genfromtxt("results/simplernn_test_loss_run1.txt", delimiter=',')[:100]

# train top 1 accuracy
train_rnn_top1acc_1 = genfromtxt("results/simplernn_train_top1acc_run1.txt", delimiter=',')[:100]
# test top 1 accuracy
test_rnn_top1acc_1 = genfromtxt("results/simplernn_test_top1acc_run1.txt", delimiter=',')[:100]

# train top 5 accuracy
train_rnn_top5acc_1 = genfromtxt("results/simplernn_train_top5acc_run1.txt", delimiter=',')[:100]

# test top 5 accuracy
test_rnn_top5acc_1 = genfromtxt("results/simplernn_test_top5acc_run1.txt", delimiter=',')[:100]
```

```python
# Plot loss 

plt.figure(figsize=(18,5))
plt.subplot(1,3,1)
plt.plot(train_rnn_loss_1, linewidth=1.5)
plt.plot(test_rnn_loss_1, linewidth=1.5)
plt.title('Rnn: Loss per epoch', fontsize = 16)
plt.xlabel('Number of epochs', fontsize = 16)
plt.ylabel('Loss value', fontsize = 16)
plt.legend(['Train loss', 'Test loss'],
           prop={'size': 14},           
            frameon=False)

plt.subplot(1, 3, 2)
plt.plot(train_rnn_top1acc_1, linewidth=1.5)
plt.plot(test_rnn_top1acc_1, linewidth=1.5)
plt.title('Rnn: Top 1 accuracy per epoch', fontsize = 16)
plt.xlabel('Number of epochs', fontsize = 16)
plt.ylabel('Top 1 accuracy', fontsize = 16)
plt.legend(['Train accuracy', 'Test accuracy'],
           prop={'size': 14},           
            frameon=False)

plt.subplot(1, 3, 3)
plt.plot(train_rnn_top5acc_1, linewidth=1.5)
plt.plot(test_rnn_top5acc_1, linewidth=1.5)
plt.title('Rnn: Top 5 accuracy per epoch', fontsize = 16)
plt.xlabel('Number of epochs', fontsize = 16)
plt.ylabel('Top 5 accuracy', fontsize = 16)
plt.legend(['Train accuracy', 'Test accuracy'],
           prop={'size': 14},           
            frameon=False)
```
